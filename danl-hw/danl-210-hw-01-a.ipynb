## Question 6

This Python code creates a tuple with seven different ages:

````python
respondent_ages = (65, 29, 25, 35, 58, 23, 19)
````

Write a Python code that uses a `for`-loop statement and an `if`-`else` chain to assign the list, `under40_list`, to `[29, 25, 35, 23, 19]` and the list, `over40_list`, to `[65, 58]`.

*Answer*

````python
respondent_ages = (65, 29, 25, 35, 58, 23, 19)

# Initialize the lists
under40_list = []
over40_list = []

# Iterate through the tuple
for age in respondent_ages:
    if age < 40:
        under40_list.append(age)
    else:
        over40_list.append(age)

under40_list, over40_list
````
